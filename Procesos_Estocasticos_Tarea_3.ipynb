```python
from numpy.lib.function_base import quantile
from numpy.core.numeric import extend_all
from scipy.special import comb
print("Bienvenido a la calculadora de probabilidades de una caminata aleatoria")
while(True):
    print("""""")
    print("""¿Que tipo de caminata aleatoria es? Escribe una opción
    1) Caminata Aleatoria Simetrica Normal
    2) Caminata Aleatoria Simetrica Pasa por Primera Vez En Un Lugar
    3) Caminata Aleatoria Asimetrica Hacia la Izquierda
    4) Caminata Aleatoria Asimetrica Hacia la Derecha
    5) Salir""")
    opcion = int(input())
    if opcion == 1:
        origen = float(input("introduce el origen:"))
        pasos = float(input("introduce los pasos:"))
        final = float(input("introduce en que lugar termina:"))
        nck = comb(pasos, (1/2)*(pasos+final), exact=False)
        prob = nck * (1/(2**pasos))
        print("La probabilidad es:", prob)
    elif opcion == 2:
        pasos = int(input("introduce los pasos:"))
        p = .5
        q = .5
        prob = p*q*(1-q)**(pasos-2)
        print("La probabilidad es:", prob)
    elif opcion == 3:
        origen = float(input("introduce el origen:"))
        pasos = float(input("introduce los pasos:"))
        q = float(input("introduce la probabilidad de q:"))
        p = 1 - q
        final = float(input("introduce en que lugar termina:"))
        nck = comb(pasos, (1/2)*(pasos+final), exact=False)
        prob = nck * (p**((1/2)*(pasos+final-origen))) * (q**((1/2)*(pasos-final+origen)))
        print("La probabilidad es:", prob)
    elif opcion == 4:
        origen = float(input("introduce el origen:"))
        pasos = float(input("introduce los pasos:"))
        p = float(input("introduce la probabilidad de q:"))
        q = 1 - q
        final = float(input("introduce en que lugar termina:"))
        nck = comb(pasos, (1/2)*(pasos+final), exact=False)
        prob = nck * (q**((1/2)*(pasos+final-origen))) * (p**((1/2)*(pasos-final+origen)))
        print("La probabilidad es:", prob)
    elif opcion == 5:
        print("")
        print("Saliendo de la calculadora")
        break
    else:
        print("comando desconocido") 
```
